# File Handling
File handling in Python is a crucial aspect of programming, as it allows you to read and write data to and from files on your computer. This can be useful for tasks like reading configuration files, processing data, or storing information.

## File Opening Modes

Modes.png

# Opening and Closing Files:
Python provides built-in functions to open and close files. You can use the **open()** function to open a file and the **close()** method to close it when you're done.

```python
# Opening a file in read mode
file = open("example.txt", 'r')

# Reading file content
content = file.read()
print(content)

# Closing the file
file.close()
```

```python
# Reading a file line by line
with open("example.txt", "r") as file:
    for line in file:
        print(line)

# Reading the entire file content
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```

# Writing to Files:
To write data to a file, open it in write mode** ('w')**. You can use **write()** or **writelines()** methods to write data.

```python
# Writing to a file
with open("output.txt", "w") as file:
    file.write("Hello, World!\n")
    file.write("This is a new line.\n")

# Writing multiple lines at once
lines = ["Line 1\n", "Line 2\n"]
with open("output.txt", "w") as file:
    file.writelines(lines)
```

```python
# Appending to a file
with open("output.txt", "a") as file:
    file.write("This is appended text.\n")
    print(file)
```

```python
import csv

# Reading CSV file
with open("data.csv", "r") as csvfile:
    csvreader = csv.reader(csvfile)
    for row in csvreader:
        print(row)

# Writing to CSV file
with open("output.csv", "w", newline="") as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(["Name", "Age"])
    csvwriter.writerow(["Alice", 25])
```
